# Ploting a boxplot of LungCapacity for smokers. Comparing the lung capacity of smokers against non-smokers.
<img src = 'https://i.ibb.co/VwWR7yq/lung-cancer.jpg' height = 500 width = 500>

```python
sns.boxplot(x=df['LungCap'],y=df['Smoke'])
plt.show()
```

##### Inference: As we can see, the youth who do not smoke have a slightly better lung capacity as compared to smokers.

# Using a Boxplot compare the 'LungCap' of males to that of females. Also compare the smokers vs non smokers 'LungCap' for males vs females

###### Checking the difference of lungs capacity between male and female.
<img src = 'https://i.ibb.co/nDg29Sh/download.png'>

```python
sns.boxplot(x=df['LungCap'],y=df['Gender'],hue=df['Smoke'])
```

##### Inference: As we can see from the above boxplots, the lung capacity of most males is higher than the lung capacity of most females. But those who do not smoke have higher lung capacity in females, whereas in males, it looks quite same.

# Using a scatterplot checking whether there is any relationship between Height(inches) and Lung Capacity. How is the relationship between the two? Checking the correlation between the two.

```python
sns.scatterplot(x=df['Height(inches)'],y=df['LungCap'])
print(df[['Height(inches)','LungCap']].corr())
```

###### Inference: There is a linear increase in Lung Capacity with respect to height of the person. Also we can see that the correlation co-efficient is +ve so there is a +ve relationship between the two.

# What is the effect of age on lung capacity? Is it the same for smokers and non smokers across the ages? Also, is it the same for males and females across ages?

```python
plt.figure(figsize=(15,7))
sns.violinplot(x=df['Age( years)'],y=df['LungCap'],hue=df['Smoke'])
plt.show()
```

##### Inference : Overall, the lung capacity increases with age. We can also see that accept for age 11 and 19, all the non smokers have a better lung capacity than smokers

```python
plt.figure(figsize=(15,7))
sns.violinplot(x=df['Age( years)'],y= df['LungCap'],hue=df['Gender'])
plt.show()
```

##### We can also see that for all age groups, males have a higher lung capacity than females.
